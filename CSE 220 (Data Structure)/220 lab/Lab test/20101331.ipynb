```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class TreeAnalyzer:
    def __init__(self, root=None):
        self.root = root

    def max_depth(self):
        def depth(root):
          if not root:
            return 0
          depth1=depth(root.left)
          depth2=depth(root.right)
          return max(depth1,depth2)+1
        return depth(self.root)

    def average_leaf_value(self):
        c=0
        sum=0

        def avg(root,c,sum):
          if not root:
            return c,sum
          if root.left and root.right:
            if isinstance(root.value,int):
              c+=1
              sum+=root.value
          if not root.left and not root.right:
            if isinstance(root.value,int):
              c+=1
              sum+=root.value
          c,sum=avg(root.left,c,sum)
          c,sum=avg(root.right,c,sum)
          return c,sum
        c,sum=avg(self.root,c,sum)
        if c==0:
          return 0
        print(sum)
        print(c)
        return sum/c
```

```python
initial_tree_root = TreeNode(30)
initial_tree_root.left = TreeNode("CAT")
initial_tree_root.right = TreeNode(9)
initial_tree_root.left.left = TreeNode(31)
initial_tree_root.left.right = TreeNode(25)
initial_tree_root.right.left = TreeNode(27)
initial_tree_root.right.right = TreeNode("DOG")

analyzer = TreeAnalyzer(initial_tree_root)
max_depth_result = analyzer.max_depth()
print(max_depth_result)

average_leaf_value_result = analyzer.average_leaf_value()
# Output the results after dynamic updates
print(f"Average Leaf Value: {average_leaf_value_result:.1f}")
```

```python
class Tree:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def tree_generation():
  root=Tree(0)
  root.left=Tree(1)
  root.right=Tree(1)
  root.left.left=Tree(2)
  root.left.right=Tree(3)
  root.right.left=Tree('None')
  root.right.right=Tree(5)
  root.right.right.left=Tree(8)
  root.right.right.right=Tree(13)
  root.right.right.right.left=Tree(21)
  root.right.right.right.right=Tree(34)
  return root

sample_tree = tree_generation()
```

```python
def print_tree(root):
    if root:

        print(root.value, end=" ")


        print_tree(root.left)
        print_tree(root.right)


print_tree(sample_tree)
```
