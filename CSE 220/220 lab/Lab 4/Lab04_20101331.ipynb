```python
class Patient:
    def __init__(self,id,name,age,bloodgroup):
        self.id=id
        self.name=name
        self.age=age
        self.bloodgroup=bloodgroup
        self.next=None
        self.prev=None

class WRM:
    def __init__(self):
        self.head=Patient(None,None,None,None)
        self.head.next=self.head
        self.head.prev=self.head

    def RegisterPatient(self,id,name,age,bloodgroup):
        new_patient=Patient(id,name,age,bloodgroup)
        new_patient.next=self.head.next
        new_patient.prev=self.head
        self.head.next.prev=new_patient
        self.head.next=new_patient

    def ServePatient(self):
        if self.head.next!=self.head:
            served_patient=self.head.next
            self.head.next=served_patient.next
            served_patient.next.prev=self.head
            return served_patient.name

    def CancelAll(self):
        self.head.next=self.head
        self.head.prev=self.head

    def CanDoctorGoHome(self):
        return self.head.next==self.head

    def ShowAllPatient(self):
        current=self.head.next
        while current!=self.head:
            print(current.name)
            current=current.next

    def ReverseTheLine(self):
        if self.head.next!=self.head:
            current=self.head.next
            while current!=self.head:
                current.next,current.prev=current.prev,current.next
                current=current.prev
            self.head.next,self.head.prev=self.head.prev,self.head.next
            return "Success."
        else:
            return "Failed."


def user():
    wrm=WRM()
    while True:
        print("\nOptions:")
        print("a. Add Patient")
        print("b. Serve Patient")
        print("c. Show All Patients")
        print("d. Can Doctor go Home?")
        print("e. Cancel all Appointments")
        print("f. Reverse The Line")

        ch=input()

        if ch=="a":
            id=int(input("ID: "))
            name=input("Name: ")
            age=int(input("Age: "))
            bloodgroup=input("BG: ")
            wrm.RegisterPatient(id, name, age, bloodgroup)
            print("Success.")
        elif ch=="b":
            result=wrm.ServePatient()
            print(f"Patient {result} is being Served.")
        elif ch=="c":
            wrm.ShowAllPatient()
        elif ch=="d":
            if wrm.CanDoctorGoHome():
                print("Yes")
            else:
                print("No")
        elif ch=="e":
            wrm.CancelAll()
            print("Success.")
        elif ch=="f":
            result=wrm.ReverseTheLine()
            print(result)
        else:
          break
user()
```
